```python
# DATASET 1 - IMAGE BLUR

import numpy as np
import matplotlib.pyplot as plt
 
# pathology, image blur
# height = [1.067022144, 0.0269186496735, 3.989729404449463]
height = [1.067022144, 0.02425217628479004, 0.022281885147094727 ,3.989729404449463]
bars = ('OpenCL', 'Cuda','Cuda Tiled', 'CPU')
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
plt.ylabel('output coding times(sec)')
plt.title("Image Blur Execution Times Dataset 1")
# Show graphic
plt.savefig("Bar_BLUR_PATH.png")
plt.show()
```
